```python
from io import StringIO ## for Python 3
import pandas as pd
import numpy as np
import cv2
import pytesseract
from pytesseract import Output
import re

pytesseract.pytesseract.tesseract_cmd  = 'C:/Program Files/Tesseract-OCR/tesseract'


path_imag = 'ejemplos_estados_financieros/temp_file.png'
```

```python
#cargar imagen
img = cv2.imread(path_imag)

gris = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
suavi = cv2.GaussianBlur(gris,(3,3),0)
#kern = np.array([[0,1,0],[1,-4,1],[0,1,0]])
#lapl = cv2.filter2D(suavi,-1,kern)
ret,Binario = cv2.threshold(suavi,40,255,cv2.THRESH_BINARY)

d = pytesseract.image_to_string(Binario, output_type=Output.DICT)

texto_pag = d['text']

print(texto_pag)
```

```python

monday = pytesseract.image_to_string(suavi, lang='spa')

#monday_lap = re.sub('\n[A-z]+\n','',monday)
monday_lap = re.sub(' ','_',monday)

mon = pd.read_csv(StringIO(monday_lap),sep='\s',lineterminator=r'\n', header=None, index_col=None)
```
